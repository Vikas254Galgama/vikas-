```python

import requests
from bs4 import BeautifulSoup as bs

url = requests.get("https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft")

soup = bs(url.content,'html.parser')

table = soup.find(class_="wikitable")

# Find All Headers
headers_rows =table.find_all("th")   
  


# Extract the first six Headers
headers = []
for header in headers_rows[0:6]:
  # Strip any extra_whitespace
  headers.append(header.text.strip())

headers

# empty_list_to store row_data
rows =[]
for row in table.find_all("tr")[1:]:
  # extract both Header both data cells
  cells = row.find_all(["th","td"])
  # clean the cell text
  row_data = [cell.text.strip() for cell in cells]
  # append data to the row
  rows.append(row_data)





row


# Extract data into a CSV File
import csv
with open('indian_military_aircraft.csv','w',newline='',encoding='utf-8') as csv_file:
  csv_writer = csv.writer(csv_file)
  csv_writer.writerow(headers)
  csv_writer.writerows(rows)

print("Table Extracted and saved to 'indian_military_aircraft.csv'.")

print('Preview:')

print(headers)

for row in rows[:]:
  print(row)  

```
